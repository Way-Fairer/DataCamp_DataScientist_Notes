# NumPy

## NumPy Arrays
Arrays are an alternative to lists where you're able to perform calculations over the mass set of values
If you have a list called height, you can convert it to an array using
np_height = np.array(height)
You can then take np_height and do a calculation like 
new_height = np_height + 2
Or you could do a short cut of 
new_height = np.array(height) + 2

Arrays can only contain one value type (Booleans, floats).
If you add two lists containing [1, 2, 3], the output would be [1, 2, 3, 1, 2, 3]
If you add two arrays that have the same contents as above, the output would be array([2, 4, 6])

## 2D NumPy Arrays
These are type numpy.ndarray, where n is n-dimensional array
2D arrays are made of a list of lists
2D arrays can subset using either way below; they do the same thing 
np_2d[0][2]
np_2d[0, 2]
A more complex way to subset would be like this. It would call a 2D array
np_2d[:, 1:2]

2D arrays enable element-wise calculations
